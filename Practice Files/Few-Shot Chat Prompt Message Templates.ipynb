```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai.chat_models import ChatOpenAI

from langchain_core.prompts import (AIMessagePromptTemplate,
                                        HumanMessagePromptTemplate,
                                        ChatPromptTemplate,
                                        FewShotChatMessagePromptTemplate)
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  seed = 365, 
                  temperature = 0,
                  max_tokens = 100)
```

```python
TEMPLATE_H = '''I've recently adopted a {pet}.
Could you suggest some {pet} names?'''
TEMPLATE_AI = '''{response}'''

message_template_h = HumanMessagePromptTemplate.from_template(template = TEMPLATE_H)
message_template_ai = AIMessagePromptTemplate.from_template(template = TEMPLATE_AI)
```

```python
example_template = ChatPromptTemplate.from_messages([message_template_h, message_template_ai])
```

```python
examples = [{'pet': 'dog',
            'response':'''  Oh, absolutely. Because I'm sure your dog is just waiting to be named 
by a sarcastic AI. How about "Bark Twain" if it's a literary hound, or "Sir Wag-a-lot" for a knightly canine? 
Maybe "Bark Obama" if it's a political pooch, or "Bark Zuckerberg" if it's into social networking. And if it's 
a small dog, "Tiny Paws of Fury" could work. You're welcome. '''},
           {'pet': 'cat',
            'response':'''  Oh, absolutely. Because I'm sure your cat is just waiting to be named 
by a sarcastic AI. How about "Furry McFarce", "Sir Meowsalot", or "Catastrophe"? '''},
           {'pet': 'fish',
            'response':'''  Oh, absolutely. Because I'm sure your fish is just waiting to be named 
            by a sarcastic AI. How about "Fin Diesel" if it's a fast swimmer, or "Gill Gates" for 
            a tech-savvy fish? Maybe "Bubbles" if it's a bubbly fish, or "Sushi" if you have a 
            sense of humor. And if it's a small fish, "Tiny Fins of Fury" could work. You're welcome. '''}]
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(examples = examples,
                                                  example_prompt = example_template,
                                                  input_variables = ['pet'])
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt, message_template_h])
```

```python
chat_value = chat_template.invoke({'pet':'rabbit'})
```

```python
chat_value
```

```python
for i in chat_value.messages:
    print(f'{i.type}: {i.content}\n')
```

```python
response = chat.invoke(chat_value)
```
